### Measure of central tendency

```python
import numpy as nd
```

```python
arr=nd.array([2,4,6,8,3])
```

```python
#mean
nd.mean(arr)
```

```python
#median
nd.median(arr)
```

```python
#mode
from scipy import stats
stats.mode(arr)
```

### Measure of diviation/spread

```python
import statistics as stats
```

```python
data=nd.array([10,20,30,40,50])
```

```python
#range
range_val=max(data)-min(data)
range_val
```

```python
#variance
variance_val = stats.variance(data)
variance_val
```

```python
#standard deviation 
std_val = nd.std(data)
std_val
```

### Measure of distribution

#### Skewness

```python
import pandas as pd
from scipy.stats import skew
data = [10, 20, 30, 40, 100] 
skew(data)
```

```python
data1 = [10, 20, 30, 40, 25] 
skew(data)
```

```python
data2 = [10, 20, 30, 40, -100] 
skew(data)
```

#### Kurtosis

```python
from scipy.stats import kurtosis
data = [10, 20, 30, 40, 100]
kurtosis(data)
```

```python
data = [10, 20, 30, 40, 20]
kurtosis(data)
```

```python
data = [10, 20, 30, 40, -500]
kurtosis(data)
```
