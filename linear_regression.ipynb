```python
import pandas as pd
```

```python
dt=pd.read_csv("Salary_dataset.csv")
dt #data_mining
```

```python
dt=dt.drop("Unnamed: 0",axis="columns")
dt #data_cleaning
```

```python
dt.info() #data_exploration
```

```python
dt.columns[dt.isna().any()]
```

```python
dt.describe() #data_exploration
```

```python
#feature engineering 
"""analyse what is feature input(independent variable) and the target output(dependent variable) , where in linear regression the x is independent (experience)
and y is dependent(salary)"""
```

```python
x=dt["YearsExperience"]
y=dt["Salary"]
print(x)
print(y)
import numpy as np
x=np.array(x).reshape(-1,1)
```

```python
#spliting the data's for training and testing 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=42)
```

```python
#training the model 
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#predicting the model is right or not using the test 
y_pred=lr.predict(x_test)
y_pred
```

```python
y_test
```

```python
#finding the slope
print(lr.coef_)
#finding the intercept 
print(lr.intercept_)
```

```python
#evaluation metrics
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
#predicitng salaries that are not in the test or train
lr.predict([[6]])
```

```python
import matplotlib.pyplot as plt
x_line = np.linspace(x.min(), x.max(), 100).reshape(-1, 1)
y_line = lr.predict(x_line)
plt.figure()
plt.scatter(x, y)
plt.plot(x_line, y_line)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary vs Years of Experience (Linear Regression)")
plt.show()
```
