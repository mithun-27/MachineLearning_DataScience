```python
p=nd.array([1,2,3,4,4.5,5])
m=nd.array([2,4,5,8,3,4])
plt.plot(p,m)
plt.title('numpy graph')
plt.xlabel('x-axis')
plt.ylabel('y-axix')
plt.show()
```

```python
plt.plot(p,m,'o')
plt.show()
```

```python
plt.plot(m,p,marker='o',ms=9)
plt.show()
```

```python
plt.plot(p,m,marker='s',mfc='yellow',mec='black',linestyle='dotted', linewidth=2, color='black')
plt.show()
```

```python
plt.scatter(m,p, color ='green')
plt.show()
```

```python
plt.bar(p,m, color='yellow')
plt.show()
```
